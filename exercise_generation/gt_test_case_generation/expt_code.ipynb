```python
def parse_java_code(java_code):
    java_code = java_code.strip()
    signature = re.search(r'(.*?){', java_code, re.DOTALL).group(1).strip()
    body_start_index = java_code.index("{") + 1
    body_end_index = java_code.rindex("}")
    body = java_code[body_start_index : body_end_index].strip()
    return signature, body
```

```python
def extract_function_signature_and_body(java_code):
    # Define a regular expression pattern to match function declarations
    pattern = r'public\s+\w+\s+\w+\s*\([^)]*\)\s*\{([^}]*)\}'

    # Search for the function declaration in the input Java code
    match = re.search(pattern, java_code)

    if match:
        # Extract the function signature and body
        function_signature = match.group(0)
        function_body = match.group(1)

        return function_signature, function_body
    else:
        return None, None
```
